```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
```

Datamızı əlavə edək

```python
df=pd.read_csv(r"C:/Users/user/Desktop/Data/Boston.csv")
```

Data haqqında məlumat alaq

```python
df.info()
```

Datamızı təsvir edək

```python
df.describe()
```

Rahatlıq üçün datamızın columnlarını  yenidən adlandıraq

```python
df.rename(columns={'crim': 'cinayətlərin_sayı', 
                   'zn': "rayonlaşmış_sahə",
                    'indus': 'qeyri_pərakəndə_biznes_nisbəti', 
                    'chas': 'Charles_dəyişəni', 
                    'nox': 'azot_konsentrasiyası', 
                    'rm': 'otaq_sayı', 
                    'age': 'yaş', 
                    'dis': 'boston_məsafə', 
                    'rad': 'magistral_yol_indeksi',
                    'tax': 'vergi', 
                    'ptratio': 'şagird_müəllim_nisbəti', 
                    'black': 'qara_dərililər', 
                    'lstat': 'əhali_aşağı_statusu', 
                    'medv': 'orta_dəyər'}, inplace=True)
```

Korrelyasiyaya baxaq

```python
df.corr()
```

Heatmap vasitəsilə baxaq

```python
plt.figure(figsize=(10,7))
sns.heatmap(df.corr(),cmap="Greens",linewidths=2)
```

```python
plt.figure(figsize=(11,7))
sns.heatmap(df.corr(),cmap='Greens',center=0,linewidths=2,annot=True)
```

Yuxarıdakı nəticəyə əsasən 1dən başqa nəzərəçarpan maximum və minimum korrelyasiyalara baxaq

```python
corr = df.corr()
maxmin = corr[((corr >= .6) | (corr <= -.6)) & (corr !=1.000)]
plt.figure(figsize=(11,7))
sns.heatmap(maxmin, cmap="Greens", linewidths=2,annot=True)
```

Ayrılıqda maximum və minimum baxaq

```python
corr = df.corr()
max = corr[corr>=(.6)]
plt.figure(figsize=(10,7))
sns.heatmap(max, cmap="Greens",linewidths=2, annot=True)
```

```python
corr = df.corr()
min = corr[corr<=-(.6)]
plt.figure(figsize=(10,7))
sns.heatmap(min, cmap="Greens",linewidths=2, annot=True)
```
